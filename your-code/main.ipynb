## Challenge 3

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

```python
A = 0.3 
B = 0.6 
AnB = 0.1 
```

* $P(\overline{A}) = 0.7$

```python
#The complement of A is 1-A.
#The probability of A not happening.
if (1-A)==0.7: 
    print(True)
else: 
    print(f"False, the right value is {1-A}")
```

* $P(A \cup B) = 0.6$

```python
#Because there is a component of dependence of A to B, which is 0.1,
#The probability of event A and B happening.

if (A+B-AnB)==0.6: 
    print(True)
else: 
    print(f"False, the right value is {round((A+B),1)}")
```

* $P(A \cap \overline{B}) = 0.2$

```python
#Probability of even A happening if B doesn't happen.
# That would be just 0.3 if it was independent, but there is an aspect that depends on B: 
#So P(AnB') = P(A)-P(AnB)
if (A-AnB)==0.2: 
    print(True)
else: 
    print(f"False, the right value is {round((A-AnB),2)}")
```

* $P(A \cup \overline{A}) = 1$

```python
#By definition, yes
if (A+(1-A))==1: 
    print(True)
else: 
    print(f"False, the right value is {round((A+(1-A)),1)}")
```

* $P(A \cap (B \cup \overline{B})) = 0.4$

```python
#The inner set is equal to 1 (B+(1-B)), and the intersection of 
innerset = 1
if (A*innerset)==0.4: 
    print(True)
else: 
    print(f"False, the right value is {round((A*innerset),1)}")
```

 $P(\overline{A} \cap \overline{B}) = 0.3$

```python
#This can be read as the probability of A and B not happenning. 
#probability not in A and not in B, so everything that would be outside A and B. 
#1-P(AuB)
if (1-(A+B-AnB))==0.3: 
    print(True)
else: 
    print(f"False, the right value is {round((1-(A+B-AnB)),1)}")
```

 $P(\overline{A \cap B}) = 0.9$

```python
#What is not in both sets.
if (1-AnB)==0.9: 
    print(True)
else: 
    print(f"False, the right value is {round((1-AnB),1)}")
```
